```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL of the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Load the dataset
titanic = pd.read_csv(url)

# Set up the plotting environment
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15, 20))

# Display the first few rows of the dataset
print(titanic.head())

# Generate a bar chart for each column
for i, column in enumerate(titanic.columns, 1):
    plt.subplot(6, 2, i)
    if titanic[column].dtype == 'object' or len(titanic[column].unique()) < 20:
        sns.countplot(y=titanic[column], palette='viridis')
        plt.title(f'Distribution of {column}', fontsize=14)
        plt.xlabel('Count')
        plt.ylabel(column)
    else:
        sns.histplot(titanic[column].dropna(), kde=False, palette='viridis', bins=20)
        plt.title(f'Distribution of {column}', fontsize=14)
        plt.xlabel(column)
        plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
